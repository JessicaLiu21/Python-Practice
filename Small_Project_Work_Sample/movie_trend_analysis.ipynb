```python
# Use genre-end point"3/genre/movie/list" to download the genre data
# define a function to connect the genre_id with genre_name
def get_genre_id(genrename):
    '''When user give a genrename, it could connect to its genre id'''
    genre_list = requests.get("https://api.themoviedb.org/3/genre/movie/list?api_key=e0dafbc8128fc7faa00fbc73e00c3b71")
    genre_list=genre_list.json()
    genre_data=json.dump(genre_list,open('genre.json','w'))
    with open("genre.json", "r") as read_file:
        genre_use = json.load(read_file)
    genre_l=genre_use['genres']
    for item in genre_l:
        if item['name']==genrename:
            return item['id']
```

```python
def release_single_month_genre(genrename,month,feature):
    '''Given the arguments of genrename,month and feature, it will return the release number of certain genre in certrain month.
       The year here is 2018'''
    genre = get_genre_id(genrename) # handle the error of different month having different days
    list1=['01','03','05','07','08','10','12']
    list2=['04','06','09','11']
    list3=['02']
    monthend=0
    if month in list1:
        monthend=31
    if month in list2:
        monthend=30
    if month in list3:
        monthend=28  #connect to "discover endpoint with date and genre name
    responses = requests.get("https://api.themoviedb.org/3/discover/movie?api_key=e0dafbc8128fc7faa00fbc73e00c3b71&page=1&primary_release_date.gte=2018-"+str(month)+"-01&primary_release_date.lte=2018-"+str(month)+"-"+str(monthend)+"&with_genres=" + str(genre))
    data = responses.json()
    if feature in data.keys(): #handling the error there is null value in the key of certain feature
        m=data[feature]
    else:
        m=0
    return m
```

```python
month=['01','02','03','04','05','06','07','08','09','10','11','12']
genre_types=['Action','Comedy','Horror','Romance','Documentary','Crime']
def all_month_genres_summary(month,genre_types,feature):
    m=[]
    types=[]
    count=[]
    for date in month:
        for genre in genre_types:
            re_num=release_single_month_genre(genre,date,feature)
            m.append(date)
            types.append(genre)
            count.append(re_num)
    data=list(zip(m,types,count))
    data=[list(row) for row in data]
    Y_Action=[]
    Y_Comedy=[]
    Y_Horror=[]
    Y_Romance=[]
    Y_Documentary=[]
    Y_Crime=[]
    for line in data:
        if line[1]=='Action':
            Y_Action.append(line[2])
        if line[1]=='Comedy':
            Y_Comedy.append(line[2])
        if line[1]=='Horror':
            Y_Horror.append(line[2])
        if line[1]=='Romance':
            Y_Romance.append(line[2])
        if line[1]=='Documentary':
            Y_Documentary.append(line[2])
        if line[1]=='Crime':
            Y_Crime.append(line[2])
    Y=[]
    Y.append(Y_Action)
    Y.append(Y_Comedy)
    Y.append(Y_Horror)
    Y.append(Y_Romance)
    Y.append(Y_Documentary)
    Y.append(Y_Crime)
    return Y
```

```python
results_summary=all_month_genres_summary(month,genre_types,'total_results')
```
